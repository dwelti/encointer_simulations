```python
import numpy as np
import matplotlib.pyplot
```

```python
def expit(x):
    y = 1 / (1 + np.exp(-x))
    return y
```

```python
# plot expit

x = np.linspace(-6, 6, 121)
y = expit(x)
matplotlib.pyplot.plot(x, y)
matplotlib.pyplot.grid()
matplotlib.pyplot.xlim(-6, 6)
matplotlib.pyplot.ylim(0, 1)
matplotlib.pyplot.xlabel('x')
matplotlib.pyplot.title('expit(x)')
matplotlib.pyplot.show()
```

```python
def find_probability_buyer(x, max_wealth):
    n=max_wealth/2
    return expit(2/n*(x-n))
```

```python
def find_probability_ceremony_participant(x, max_wealth):
    scale_value=max_wealth/2
    translate_value=max_wealth/1.5
    return 1-expit(6/scale_value*(x-translate_value))
```

```python
# plot Probability Distribution of Purchasing Products

x = np.linspace(0, 1000, 121)
y = find_probability_buyer(x,1000)
matplotlib.pyplot.plot(x, y)
matplotlib.pyplot.grid()
matplotlib.pyplot.xlim(0, 1000)
matplotlib.pyplot.ylim(0, 1)
matplotlib.pyplot.xlabel('Wealth of Agent [LEU+CHF*XR]')
matplotlib.pyplot.ylabel('Probability of Agent to Purchase Products')
matplotlib.pyplot.title('Probability Distribution of Purchasing Products')
matplotlib.pyplot.show()
matplotlib.pyplot.savefig('Probability Distribution of Purchasing Products', dpi=300)
```

```python
# plot Probability Distribution of Attending Ceremony

x = np.linspace(0, 1000, 121)
y = find_probability_ceremony_participant(x,1000)
matplotlib.pyplot.plot(x, y)
matplotlib.pyplot.grid()
matplotlib.pyplot.xlim(0, 1000)
matplotlib.pyplot.ylim(0, 1)
matplotlib.pyplot.xlabel('Wealth of Agent [LEU+CHF*XR]')
matplotlib.pyplot.ylabel('Probability of Agent Attending Ceremony')
matplotlib.pyplot.title('Probability Distribution of Attending Ceremony')
matplotlib.pyplot.show()
matplotlib.pyplot.savefig('Probability Distribution of Attending Ceremony', dpi=300)
```
